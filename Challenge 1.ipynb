### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación
Ingresos totales por tienda

En este análisis se calcula el ingreso total de cada tienda sumando los valores de la columna precio.


```python
tiendas = [tienda1, tienda2, tienda3, tienda4]
ingresos = []
for i, tienda in enumerate(tiendas, start=1):
    ingreso = tienda['Precio'].sum()
    ingresos.append(ingreso)
    print(f"Ingreso total Tienda {i}: {ingreso}")
ingresos
```

# 2. Ventas por categoría
En esta sección se analiza cuáles son las categorías de productos más vendidos en cada tienda.

```python
categorias_tiendas = []
for i, tienda in enumerate(tiendas, start=1):
    categorias = tienda['Categoría del Producto'].value_counts()
    categorias_tiendas.append(categorias)
    print(f"\nCategorías más populaes de la Tienda {i}:\n\n{categorias}")
```

# 3. Calificación promedio de la tienda
Este análisis permite conocer el nivel de satisfacción de los clientes en cada tienda.

```python
calificaciones_promedio = []
for i, tienda in enumerate(tiendas, start=1):
    promedio = tienda['Calificación'].mean()
    calificaciones_promedio.append(promedio)
    print(f"Tienda {i}: {promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
mas_vendidos = {}
menos_vendidos = {}
for i, tienda in enumerate(tiendas, start=1):
    productos = tienda['Producto'].value_counts()
    mas_vendidos[f"Tienda {i}"] = productos.head()
    menos_vendidos[f"Tienda {i}"] = productos.tail()
    print(f"\n      Productos Tienda {i}")
    print("\nProductos más vendidos:")
    print(productos.head())
    print("\nProductos menos vendidos:")
    print(productos.tail())
```

# 5. Valor promedio del envío por tienda

```python
envio_promedio = {}
for i, tienda in enumerate(tiendas, start=1):
        promedio = tienda['Costo de envío'].mean()
        envio_promedio[f"Tienda {i}"] = promedio
        print(f"Valor promedio de envío en Tienda {i}: {promedio:.2f}")
```

#Visualización

```python
#Ingresos por tienda
tiendas_nombres=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
plt.figure(figsize=(8,5))
plt.bar(tiendas_nombres, ingresos, color='skyblue')
plt.title("Ingresos totales por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))
fig.suptitle("Distribución de Categorías por Tienda", fontsize=16, fontweight='bold')
for i, categorias in enumerate(categorias_tiendas):
    fila = i // 2
    col = i % 2
    axes[fila, col].pie(categorias, labels=categorias.index, autopct='%1.1f%%')
    axes[fila, col].set_title(f"Tienda {i+1}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(tiendas_nombres, calificaciones_promedio, marker='o', linewidth=2, color='purple')
plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Promedio de calificación")
plt.grid(True)
plt.show()
```

```python
#Productos más Vendidos
plt.figure(figsize=(12,8))
for i, (tienda, datos) in enumerate(mas_vendidos.items(), 1):
    plt.subplot(2,2,i)
    datos = datos.sort_values()
    ax = datos.plot(kind='barh', color='lightgreen')
    plt.title(f"Productos más vendidos {tienda}")
    plt.xlabel("Cantidad vendida")
    plt.ylabel("Producto")
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    # Valores al final de cada barra
    for index, value in enumerate(datos):
        plt.text(value, index, f" {value}", va='center')
plt.suptitle("Productos más vendidos por tienda", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Productos menos Vendidos
plt.figure(figsize=(12,8))
for i, (tienda, datos) in enumerate(menos_vendidos.items(), 1):
    plt.subplot(2,2,i)
    ax = datos.plot(kind='barh', color='salmon')
    plt.title(f"Productos menos vendidos {tienda}")
    plt.xlabel("Cantidad vendida")
    plt.ylabel("Producto")
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    # Valores al final de cada barra
    for index, value in enumerate(datos):
        plt.text(value, index, f" {value}", va='center')
plt.suptitle("Productos menos vendidos por tienda", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df_mas = pd.DataFrame(mas_vendidos).fillna(0)
df_menos = pd.DataFrame(menos_vendidos).fillna(0)
df_mas['Total'] = df_mas.sum(axis=1)
df_mas = df_mas.sort_values('Total', ascending=False).drop(columns='Total')

df_menos['Total'] = df_menos.sum(axis=1)
df_menos = df_menos.sort_values('Total', ascending=True).drop(columns='Total')
# Graficar productos más vendidos
df_mas.plot(kind='bar', figsize=(14,6))
plt.title("Comparación de Productos Más Vendidos por Tienda")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Tiendas")
plt.tight_layout()
plt.show()
# Graficar productos menos vendidos
df_menos.plot(kind='bar', figsize=(14,6))
plt.title("Comparación de Productos Menos Vendidos por Tienda")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Tiendas")
plt.tight_layout()
plt.show()
```

```python
df_mas = pd.DataFrame(mas_vendidos).fillna(0)
df_menos = pd.DataFrame(menos_vendidos).fillna(0)
# Sumar la cantidad total de productos más vendidos por tienda
total_mas = df_mas.sum()
# Sumar la cantidad total de productos menos vendidos por tienda
total_menos = df_menos.sum()
# Mostrar los totales
print("Total de Productos Más Vendidos por Tienda:")
print(total_mas)
print("\nTotal de Productos Menos Vendidos por Tienda:")
print(total_menos)
```

```python
#Valor Promedio de Envio
envio_series = pd.Series(envio_promedio)
plt.figure(figsize=(8,5))
plt.plot(envio_series.index, envio_series.values, marker='o', linewidth=2,color='green' )
for i, valor in enumerate(envio_series.values):
    plt.text(i, valor, f"{valor:.2f}", ha='center', va='bottom')
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo promedio de envío")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

# Informe de Análisis de Tiendas

## Introducción
El objetivo de este análisis es determinar **cuál de las cuatro tiendas evaluadas representa el menor desempeño**, para que el Sr. Juan pueda considerar venderla e invertir en un nuevo negocio.  
Para ello, se analizaron cinco aspectos fundamentales: ingresos totales, categorías de productos más populares, calificación promedio de clientes, productos más y menos vendidos, y costo promedio de envío. Este informe integra los resultados de cada uno de estos análisis y proporciona una recomendación final basada en datos.
## Desarrollo
### 1. Ingresos totales por tienda

Se calcularon los ingresos sumando los precios de todos los productos vendidos en cada tienda. Los resultados muestran diferencias significativas en el desempeño económico:

- **Tienda 1:** 1.150.880.400  
- **Tienda 2:** 1.116.343.500  
- **Tienda 3:** 1.098.019.600  
- **Tienda 4:** 1.038.375.700  

Se observa que la **Tienda 4** genera los menores ingresos totales, mientras que la **Tienda 1** es la que más factura.

---

### 2. Categorías de productos más y menos vendidas

El análisis por categoría revela que, en todas las tiendas, la categoría **Muebles** es la más popular, seguida de **Electrónicos** y **Juguetes**, mientras que categorías como **Artículos para el hogar** y **Instrumentos musicales** presentan menor venta. Las tiendas tienen una distribución parecida en las categorías, sin embargo la **Tienda 4** aunque tiene porcentajes parecidos de ventas puede representar una dificultad que tiene categorías diferentes a las otras tiendas lo que dificultaria la administración general, ya que a esta tienda se le debe manejar cantidad de producto diferenete a las otras tiendas.

---

### 3. Calificación promedio de los clientes

El nivel de satisfacción de los clientes se evaluó mediante la calificación promedio:

- **Tienda 1:** 3.98  
- **Tienda 2:** 4.04  
- **Tienda 3:** 4.05  
- **Tienda 4:** 4.00  

Si bien la Tienda 3 tiene la mejor calificación promedio, la Tienda 4 presenta un desempeño ligeramente inferior a las demás, indicando un nivel de satisfacción adecuado pero no destacado. Y aunque la Tienda 1 tendria la peor calificación no esta muy alejada de la calificación mayor.

---

### 4. Productos más y menos vendidos

Los productos más vendidos varían entre tiendas, destacando artículos de tecnología y muebles como los líderes en ventas. Por otro lado, productos con menor venta incluyen libros especializados, ciertos instrumentos musicales y electrodomésticos menos populares. La Tienda 2 concentra más ventas en sus productos estrella, la Tienda 1 mantiene un desempeño sólido y equilibrado, la Tienda 3 tiene menor fuerza en sus productos más vendidos, y la Tienda 4 muestra ventas moderadas tanto en productos top como en los menos populares, reflejando dispersión en sus ventas. Por tanto,Tienda 3 refleja un rendimiento más débil en productos clave lo que seria problematico.

---

### 5. Costo promedio de envío

El análisis del costo de envío promedio muestra:

- **Tienda 1:** 26.018,61  
- **Tienda 2:** 25.216,24  
- **Tienda 3:** 24.805,68  
- **Tienda 4:** 23.459,46  

La Tienda 4 presenta el **costo de envío más bajo**, lo que podría indicar un margen menor por venta y un menor aporte a la rentabilidad general.

---

## Conclusión y recomendación

Considerando todos los aspectos analizados, ingresos totales, categorías y productos vendidos, calificación promedio y costo de envío, se concluye que la **Tienda 4** es la que presenta el menor rendimiento global. A pesar de mantener una calificación de clientes razonable, sus ingresos son los más bajos y aunque no muestra productos con ventas reducidas, tiene un menor costo promedio de envío que podría reflejar márgenes más ajustados. Además, tiene categorias de más ventas diferentes a las de los otros locales lo cual haría que comprar inventario sea menos eficiente para un  solo dueño.

Por lo tanto, se recomienda al **Sr. Juan** considerar la venta de la **Tienda 4** para reinvertir en un nuevo negocio, permitiendo optimizar recursos y enfocarse en tiendas con mejor desempeño económico y mayor potencial de crecimiento.
